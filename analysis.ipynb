```python
from constructor import TextProcessor
import matplotlib.pyplot as plt
```

```python
# Creates an instance of TextProcessor and calls the DF cretor function to return
# a dataframe with the word counts in each article
df = TextProcessor('text.txt').df_creator()

# Shows the structure of the dataframe
df.head(10)
```

```python
# Adds a column with the total amount of times a word was counted in all articles
df['total'] = df.sum(axis = 1)

# Checks the dataframe structure again to see the new column
df.head(10)
```

```python
# Plots the number of times a word appears in the entire collection versus the frequency of that count
plt.style.use('ggplot')
count_freq = df.groupby('total').total.count().plot(kind = 'bar',
                                            figsize = (10,10),
                                            fontsize = 12,
                                            width = 0.8)

count_freq.set_title('Count Distribution of Words in all Articles', fontsize = 18)
count_freq.set_xlabel('Number of times a word appears in the entire collection', fontsize = 16)
count_freq.set_ylabel('Frequency of each count', fontsize = 16)
```

```python
# plot the most common words in each article
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize = (10,7))
fig.suptitle('Most Frequent Words Per Article', fontsize = 20)
ax1.bar(df.article_1.sort_values(ascending = False)[:5].index, df.article_1.sort_values(ascending = False)[:5].values)
ax2.bar(df.article_2.sort_values(ascending = False)[:5].index, df.article_2.sort_values(ascending = False)[:5].values)
ax3.bar(df.article_3.sort_values(ascending = False)[:5].index, df.article_3.sort_values(ascending = False)[:5].values)

ax2.set_ylabel('Word Counts', fontsize = 16)
ax3.set_xlabel('Most Common Words', fontsize = 16)
```

```python
# create a df with only the words with more than 4 letters
df_filtered = df[[True if len(word) > 4 else False for word in df.index.values]]

# plot the most common words in each article bigger than 4 letters
figf, (ax1f, ax2f, ax3f) = plt.subplots(3, figsize = (10,7))
figf.suptitle('Most Frequent Words Per Article with more than 4 letters', fontsize = 20)
ax1f.bar(df_filtered.article_1.sort_values(ascending = False)[:5].index, df_filtered.article_1.sort_values(ascending = False)[:5].values)
ax2f.bar(df_filtered.article_2.sort_values(ascending = False)[:5].index, df_filtered.article_2.sort_values(ascending = False)[:5].values)
ax3f.bar(df_filtered.article_3.sort_values(ascending = False)[:5].index, df_filtered.article_3.sort_values(ascending = False)[:5].values)

ax2f.set_ylabel('Word Counts', fontsize = 16)
ax3f.set_xlabel('Most Common Words', fontsize = 16)
```
